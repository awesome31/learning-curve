# Derivatives

Derivatives are the instantaneous rate of change of a function. Let us try and plot a simple graph.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(x, y, x_label, y_label, title):
    # Plot the function
    plt.plot(x, y)
    # Add labels and title
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    # Add a legend
    plt.legend()
    plt.show()
```

```python
def f(x):
    return x ** 2

x = np.linspace(0, 10, 100)
y = f(x)

plot(x, y, "x", "X square", "Plot")
```

```python
x_dx = x + 0.01

y_dx = ((f(x_dx) - f(x)) / x_dx)
plot(x, y_dx, "x", "X square", "Plot")
```
